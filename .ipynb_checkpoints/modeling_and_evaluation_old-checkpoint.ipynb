## Import Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
import time
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

**Loading the Clean 30 mph DataFrame from CSV**

```python
df_cleaned = pd.read_csv(f'data/30mph_accident_clean_data.csv', low_memory=False)
```

```python
df_cleaned.info()
```

## Feature Engineering

```python
df_cleaned['casualty_vehicle_ratio'] = df_cleaned['number_of_casualties'] / df_cleaned['number_of_vehicles']
```

```python
df_cleaned['weather_road_interaction'] = df_cleaned['weather_conditions'].astype(str) + "_" + df_cleaned['road_surface_conditions'].astype(str)
```

### Encoding Categorical Variables

```python
df_cleaned['accident_severity'].value_counts()
```

### Cyclic Encoding

```python
day_map = {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}
df_cleaned['day_of_week'] = df_cleaned['day_of_week'].map(day_map)
df_cleaned['day_of_week_sin'] = np.sin(2 * np.pi * df_cleaned['day_of_week'] / 7)
df_cleaned['day_of_week_cos'] = np.cos(2 * np.pi * df_cleaned['day_of_week'] / 7)
df_cleaned.drop('day_of_week', axis=1, inplace=True)
```

#### Label Encoding

```python
severity_map = {'Slight':0,'Serious':1, 'Fatal':2}
light_Conditions_map = {'Daylight': 1, 'Dark_Lit': 2, 'Dark_Unknown': 3, 'Dark_unlit': 4, 'Dark_No_Light': 5}
area_map = {'Urban':-0.5, 'Rural':0.5}
```

```python
df_cleaned['accident_severity'] = df_cleaned['accident_severity'].map(severity_map)
df_cleaned['light_conditions'] = df_cleaned['light_conditions'].map(light_Conditions_map)
df_cleaned['urban_or_rural_area'] = df_cleaned['urban_or_rural_area'].map(area_map)
```

#### One Hot Encoding

```python
df_cleaned = pd.get_dummies(df_cleaned, columns=['junction_control', 'junction_detail', 'road_surface_conditions',
             'road_type', 'weather_conditions', 'age_band_of_driver', 'vehicle_type', 'weather_road_interaction'], dtype=int)
```

```python
df_cleaned.head()
```

## Feature Selection

```python
def drop_low_correlated_columns(df, threshold=0.4):
    # Calculate the correlation matrix
    corr_matrix = df.corr().abs()

    # Get the columns to drop
    columns_to_drop = []
    for i, col in enumerate(corr_matrix.columns):
        for j in range(i + 1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) >= threshold:
                columns_to_drop.append(col)
                break

    # Drop the columns and return the modified dataframe
    return df.drop(columns=columns_to_drop, axis=1)
```

```python
def drop_collinear_columns(df, threshold=0.8):
    # Compute the correlation matrix
    corr_matrix = df.corr().abs()

    # Create an upper triangle matrix to avoid double-counting pairs
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    # Identify columns to drop (where correlation exceeds the threshold)
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]

    # Drop the collinear columns and return the modified dataframe
    return df.drop(columns=to_drop)
```

### Splitting the Data into Training and Test Sets

```python
X = drop_collinear_columns(drop_low_correlated_columns(df_cleaned.drop(['accident_severity', 'accident_year','date', 'time'], axis=1)))
y = df_cleaned['accident_severity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
                                                                                                                                                                                                                    
```

```python
y_train.value_counts()
```

```python
X_train.info()
```

### Resampling to Handle Class Imbalance

```python
# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to resample the training data
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
y_train_resampled.value_counts()
```

```python
# Custom colors for each class
palette = {'0': '#3f9d7f', '1': '#FF8C00', '2': '#FF0000'}  

# Create side-by-side subplots
fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# Plot the distribution of classes in y_train (Before SMOTE)
sns.countplot(x=y_train, palette=palette, ax=axes[0])
axes[0].set_title("Class Distribution (Before SMOTE)")
axes[0].set_xlabel("Class")
axes[0].set_ylabel("Count")
axes[0].grid(False)

# Plot the distribution of classes in y_train_resampled (After SMOTE)
sns.countplot(x=y_train_resampled, palette=palette, ax=axes[1])
axes[1].set_title("Class Distribution (After SMOTE)")
axes[1].set_xlabel("Class")
axes[1].set_ylabel("Count")
axes[1].grid(False)

# Display the plots
plt.tight_layout()
plt.show()
```

### Standard Scaler

```python
standard_scaler = StandardScaler()
X_train_standard = standard_scaler.fit_transform(X_train)
X_test_standard = standard_scaler.transform(X_test)
```

## Building Models

```python
def evaluate_model(model, X, y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=1)
    conf_matrix = confusion_matrix(y_test, y_pred)

    print(f"Accuracy: {accuracy * 100:.2f}%\n")
    print(f"F1 Score: {f1 * 100:.2f}%")
    print("\nClassification Report:\n", classification_report(y_true, y_pred, zero_division=1))

    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='GnBu')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```

## Bagging (Bootstrap Aggregating) Models

### Bagged Decision Tree

```python
model = BaggingClassifier(estimator=DecisionTreeClassifier(max_depth=5), n_estimators=300, random_state=42, n_jobs=-1)
model.fit(X_train_standard[:10000], y_train_resampled[:10000])

# Make predictions on the test data
y_pred = model.predict(X_test_standard)

# Evaluate the model
evaluate_model(model, X_train_standard, y_test, y_pred)
```

### Random Forest

```python
model = RandomForestClassifier(n_estimators=300, random_state=42)
model.fit(X_train_standard[:10000], y_train_resampled[:10000])

# Make predictions on the test data
y_pred = model.predict(X_test_standard)

# Evaluate the model
evaluate_model(model,X_train_standard,y_test,y_pred)
```

```python
feature_names = X.columns 

# Get feature importances from the model
feature_importances = model.estimators_[0].feature_importances_

# Create a DataFrame for feature importance and sort them
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
})

# Sort features by importance in descending order and get the top 10
top_10_features = feature_importance_df.sort_values(by='Importance', ascending=False).head(10)

# Plot the top 10 feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=top_10_features)
plt.title('Top 10 Feature Importance from Random Forest Model', fontsize=14)
plt.tight_layout()
plt.show()
```

## Boosting Models:

### Gradient Boosting

```python
model = GradientBoostingClassifier(n_estimators=300, learning_rate=0.8, loss='log_loss', random_state=42)
model.fit(X_train_standard[:10000], y_train_resampled[:10000])

# Make predictions on the test data
y_pred = model.predict(X_test_standard)

# Evaluate the model
evaluate_model(model,X_train_standard,y_test,y_pred)
```

### XGBoost

```python
# Initialize the XGBoost classifier
model = xgb.XGBClassifier(objective='multi:softmax', eval_metric='mlogloss', random_state=42)
model.fit(X_train_standard[:10000], y_train_resampled[:10000])

# Make predictions on the test data
y_pred = model.predict(X_test_standard) 

# Evaluate the model using the custom evaluate_model function
evaluate_model(model, X_train_standard, y_test, y_pred)
```

```python
# Get feature importance scores from the model
importance = model.feature_importances_

# Create list of feature names (use the original feature names from your dataset)
feature_names = X_train.columns  # or list(X_train.columns)

# Create DataFrame for feature importance
feature_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importance
})

# Sort by importance value in descending order
feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=feature_importance_df.head(10),
    x='Importance',
    y='Feature',
    color='steelblue'
)

# Customize the plot
plt.title('Top 10 Feature Importance from XGBoost Model', fontsize=14, pad=20)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')

# Add grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

# Print the importance values
print("\nFeature Importance Values:")
print(feature_importance_df.head(10).to_string(index=False))
```

### Perform Cross Validation

```python
# Define the classifiers
classifiers = {
    'Decision Tree': BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=300, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=300, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=300, learning_rate=0.8, loss='log_loss', random_state=42),
    'XGB': xgb.XGBClassifier(objective='multi:softmax', eval_metric='mlogloss', random_state=42)
}

# Initialize an empty list to store the results
results = []

# Perform cross-validation for each classifier
for name, clf in classifiers.items():
    print(f"Evaluating {name}...")
    
    # Start the timer
    start_time = time.time()
    
    # Perform cross-validation
    cv_accuracies = cross_val_score(clf, X_train_standard[:10000], y_train_resampled[:10000], cv=5, scoring='accuracy')
    cv_f1_scores = cross_val_score(clf, X_train_standard[:10000], y_train_resampled[:10000], cv=5, scoring='f1_weighted')
    
    # Fit the model on the training data and predict on the test data
    clf.fit(X_train_standard[:10000], y_train_resampled[:10000])
    y_pred = clf.predict(X_test_standard)
    
    # Evaluate the classifier
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)

    # Calculate the elapsed time
    end_time = time.time()
    elapsed_time = end_time - start_time

    # Store the results
    results.append({
        'Classifier': name,
        'Cross-Validation Accuracy': cv_accuracies.mean(),
        'Cross-Validation F1 Score': cv_f1_scores.mean(),
        'Test Set Accuracy': accuracy,
        'Test Set F1 Score': f1,
        'Time (s)': elapsed_time
    })

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Display the results
print("\nCross-Validation Results:")
results_df
```

### Evaluating Model Performance on Cross Validation

```python
# Plot 1: Bar plot comparing all metrics
plt.figure(figsize=(12, 6))
metrics = ['Cross-Validation Accuracy', 'Cross-Validation F1 Score', 
           'Test Set Accuracy', 'Test Set F1 Score']
df_melted = pd.melt(results_df, id_vars=['Classifier'], value_vars=metrics)
sns.barplot(x='Classifier', y='value', hue='variable', data=df_melted)
plt.xticks(rotation=45)
plt.title('Model Performance Comparison')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Execution Time Comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='Classifier', y='Time (s)', data=results_df)
plt.xticks(rotation=45)
plt.title('Model Execution Time Comparison')
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
# Plot 3: Cross-validation vs Test Performance
plt.figure(figsize=(12, 6))
width = 0.35
x = range(len(results_df['Classifier']))
plt.bar(x, results_df['Cross-Validation Accuracy'], width, label='CV Accuracy')
plt.bar([i + width for i in x], results_df['Test Set Accuracy'], width, label='Test Accuracy')
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.title('Cross-validation vs Test Set Performance')
plt.xticks([i + width/2 for i in x], results_df['Classifier'], rotation=45)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Heatmap of performance metrics
plt.figure(figsize=(10, 6))
metrics = ['Cross-Validation Accuracy', 'Cross-Validation F1 Score', 
           'Test Set Accuracy', 'Test Set F1 Score']
sns.heatmap(results_df[metrics].T, annot=True, fmt='.3f', cmap='YlOrRd',
            xticklabels=results_df['Classifier'], yticklabels=metrics)
plt.title('Performance Metrics Heatmap')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
