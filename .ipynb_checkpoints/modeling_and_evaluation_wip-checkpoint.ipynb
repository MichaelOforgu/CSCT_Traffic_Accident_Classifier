```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

def process_features_with_pca(df, corr_threshold=0.4, pca_components=None):
    """
    Drops low-correlated columns and applies PCA to reduce dimensionality.
    
    Args:
        df (pd.DataFrame): Input dataframe with features.
        corr_threshold (float): Threshold for dropping highly correlated columns.
        pca_components (int or float): Number of PCA components to retain. If float (0 < pca_components < 1), it retains that proportion of variance.
        
    Returns:
        pd.DataFrame: Transformed dataframe with reduced features.
    """
    # Step 1: Drop collinear columns based on correlation threshold
    corr_matrix = df.corr().abs()
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > corr_threshold)]
    df_reduced = df.drop(columns=to_drop)

    print(f"Columns dropped due to high correlation: {to_drop}")
    
    # Step 2: Standardize the data after dropping highly correlated columns
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df_reduced)
    
    # Step 3: Apply PCA to further reduce dimensionality
    if isinstance(pca_components, float) and 0 < pca_components < 1:
        # If pca_components is a float, use it to specify the proportion of variance to retain
        pca = PCA(n_components=pca_components)
    else:
        # Otherwise, use pca_components as the number of components
        pca = PCA(n_components=pca_components)

    pca_result = pca.fit_transform(X_scaled)
    
    # Create a DataFrame from PCA results
    X_pca_df = pd.DataFrame(
        pca_result, 
        columns=[f"PC{i+1}" for i in range(pca_result.shape[1])],
        index=df.index
    )
    
    print(f"Explained variance by each component: {pca.explained_variance_ratio_}")
    print(f"Total explained variance: {sum(pca.explained_variance_ratio_)}")
    
    return X_pca_df

# Example usage:

# Drop unnecessary columns from the dataframe
X = df_cleaned.drop(['accident_severity', 'accident_year', 'date', 'time'], axis=1)
y = df_cleaned['accident_severity']

# Apply the PCA function with correlation threshold and components
X_pca_df = process_features_with_pca(X, corr_threshold=0.4, pca_components=0.95)  # Adjust pca_components as needed
```

```python
def select_features_based_on_correlation(X, threshold=1.2):
    """
    Select features based on correlation matrix, dropping highly correlated features.
    
    Parameters:
    X (pd.DataFrame): The input features (dataframe).
    threshold (float): Correlation threshold to identify highly correlated features (default is 0.8).
    
    Returns:
    pd.DataFrame: A dataframe with highly correlated features dropped.
    """
    
    # Compute the correlation matrix and take the absolute values
    corr_matrix = X.corr().abs()

    # Identify pairs of highly correlated features
    high_corr_pairs = [(i, j) for i in range(len(corr_matrix)) for j in range(i+1, len(corr_matrix)) if corr_matrix.iloc[i, j] > threshold]
    
    # List of features to drop
    to_drop = []
    
    # Loop through the high correlation pairs and drop one feature from each pair
    for i, j in high_corr_pairs:
        to_drop.append(X.columns[i])  # Drop the first feature in the pair (can modify logic here if needed)
    
    # Drop the selected features and return the modified dataframe
    X_selected = X.drop(columns=set(to_drop))
    
    return X_selected
```

```python
def drop_low_correlated_columns(df, threshold=0.4):
    # Calculate the correlation matrix
    corr_matrix = df.corr().abs()

    # Get the columns to drop
    columns_to_drop = []
    for i, col in enumerate(corr_matrix.columns):
        for j in range(i + 1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) >= threshold:
                columns_to_drop.append(col)
                break

    # Drop the columns and return the modified dataframe
    return df.drop(columns=columns_to_drop, axis=1)
```

```python
def drop_collinear_columns(df, threshold=0.8):
    # Compute the correlation matrix
    corr_matrix = df.corr().abs()

    # Create an upper triangle matrix to avoid double-counting pairs
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    # Identify columns to drop (where correlation exceeds the threshold)
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]

    # Drop the collinear columns and return the modified dataframe
    return df.drop(columns=to_drop)
```

```python
def process_features_with_pca(df, corr_threshold=0.4, pca_components=None):
   corr_matrix = df.corr().abs()
   upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
   to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > corr_threshold)]
   df_reduced = df.drop(columns=to_drop)
   print(f"Columns dropped due to high correlation: {to_drop}")
   
   pca = PCA(n_components=pca_components)
   pca_result = pca.fit_transform(df_reduced)
   
   pca_df = pd.DataFrame(
       pca_result, 
       columns=[f"PC{i+1}" for i in range(pca_result.shape[1])],
       index=df.index
   )
   
   print(f"Explained variance by each component: {pca.explained_variance_ratio_}")
   return pca_df, pca
```

```python
def analyze_pca_features(original_df_before_pca, pca, df_reduced):
   feature_names = df_reduced.columns  # Use reduced dataframe columns
   components_df = pd.DataFrame(
       abs(pca.components_),
       columns=feature_names,
       index=[f'PC{i+1}' for i in range(len(pca.components_))]
   )
   
   for pc in components_df.index:
       print(f"\n{pc} top contributing features:")
       loadings = components_df.loc[pc].sort_values(ascending=False)
       print(loadings.head())
       
   return components_df
```

### Perform Cross Validation

```python
# Define the classifiers
classifiers = {
    'Decision Tree': BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=300, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=300, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=300, learning_rate=0.8, loss='log_loss', random_state=42),
    'XGB': xgb.XGBClassifier(objective='multi:softmax', eval_metric='mlogloss', random_state=42)
}

# Initialize an empty list to store the results
results = []

# Perform cross-validation for each classifier
for name, clf in classifiers.items():
    print(f"Evaluating {name}...")
    
    # Start the timer
    start_time = time.time()
    
    # Perform cross-validation
    cv_accuracies = cross_val_score(clf, X_train_standard[:10000], y_train_resampled[:10000], cv=5, scoring='accuracy')
    cv_f1_scores = cross_val_score(clf, X_train_standard[:10000], y_train_resampled[:10000], cv=5, scoring='f1_weighted')
    
    # Fit the model on the training data and predict on the test data
    clf.fit(X_train_standard[:10000], y_train_resampled[:10000])
    y_pred = clf.predict(X_test_standard)
    
    # Evaluate the classifier
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)

    # Calculate the elapsed time
    end_time = time.time()
    elapsed_time = end_time - start_time

    # Store the results
    results.append({
        'Classifier': name,
        'Cross-Validation Accuracy': cv_accuracies.mean(),
        'Cross-Validation F1 Score': cv_f1_scores.mean(),
        'Test Set Accuracy': accuracy,
        'Test Set F1 Score': f1,
        'Time (s)': elapsed_time
    })

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Display the results
print("\nCross-Validation Results:")
results_df
```
